```python
# import
%matplotlib inline
import matplotlib.pyplot as plt
from pylab import *
import numpy as np
import math
import time
import cvxopt
from cvxopt import matrix
from cvxopt import solvers
from scipy.spatial.distance import cdist
from sklearn.datasets import make_classification
from sklearn.datasets import make_moons
```

```python
# K_Fold
def K_Fold(n, n_folds=10, shuffle=False):
    step = n // n_folds
    remainder = n % n_folds
    
    # generate fold sizes
    fold_sizes = (step) * np.ones(n_folds, dtype=np.int)
    fold_sizes[:remainder] += 1
    
    train_idx = []
    test_idx = []
    
    sequence = np.array(range(n))
    if shuffle:
        np.random.shuffle(sequence)
        
    cursor = 0
    for fs in fold_sizes:
        test_fold = sequence[cursor:cursor + fs]
        test_idx.append(test_fold)
        train_fold = np.delete(sequence, range(cursor, cursor + fs))
        train_idx.append(train_fold)
        cursor += fs
    return train_idx, test_idx
```

```python
# fit transform
def fit_transform(X, degree):
#     Z = np.insert(X, 0, 1, axis=1)
    if degree > 1:
        for i in range(2, degree + 1):
            Z = np.append(Z, X ** i, axis=1)
    return Z
```

```python
# fit
def fit(X, y, c=1., sigma=0.5, degree=1, sov_lower=0.001, mode=None, kernel_func=None):
    ker = None
    if kernel_func == None:
        ker = np.dot(X, X.T)
    elif kernel_func == 'poly':
        ker = (np.dot(X, X.T) + 1) ** degree
    elif kernel_func == 'gaus':
        ker = np.exp(-0.5 * np.square(cdist(X, X, 'euclidean')) / sigma ** 2)
    
    P = np.dot(y, y.T) * ker
    P = matrix(P)
    q = -1. * np.ones([X.shape[0], 1])
    q = matrix(q)
    
    G = -1. * np.eye(X.shape[0])
    h = 1. * np.zeros([X.shape[0], 1])
    if mode == 'soft':
        G = np.insert(G, 0, np.eye(X.shape[0]), axis=0)
        h = np.insert(h, 0, c * np.ones([X.shape[0], 1]), axis=0)
    G = matrix(G)
    h = matrix(h)
    
    A = 1. * y.T
    A = matrix(A)
    
    b = matrix(0.)
    
    # 1. calculate alpha
    sol = solvers.qp(P, q, G, h, A, b)
    alpha = np.array(sol['x'])
    
    # 2. get sv set
    alpha_sv = alpha[[i for i in range(alpha.shape[0]) if alpha[i] > sov_lower]]
    X_sv = X[[i for i in range(alpha.shape[0]) if alpha[i] > sov_lower], :]
    y_sv = y[[i for i in range(alpha.shape[0]) if alpha[i] > sov_lower]]
    
    # 3. calculate W and W_0
    W = np.zeros([1, X_sv.shape[1]]) # (1, #sv)
    for i in range(X_sv.shape[0]):
        W += alpha_sv[i] * y_sv[i] * np.expand_dims(X_sv[i], axis=0)
        
    W_0 = 0. # scalar
    for i in range(X_sv.shape[0]):
        W_0 += y_sv[i] - np.dot(W, np.expand_dims(X_sv[i], axis=1))
    W_0 /= X_sv.shape[0]
    
    return alpha_sv, X_sv, y_sv, alpha, W, W_0
```

```python
def predict(X, W, W_0):
    y_hat = (np.dot(W, X.T) + W_0).T
    y_hat = (y_hat > 0) + 0
    return y_hat
```

```python
# score
def score(X, y):
    c_matrix = np.zeros((2, 2))
    for j in range(X.shape[0]):
        c_matrix[int(X[j])][int(y[j])] += 1
    
    accuracy = 1. * np.sum(c_matrix[i][i] for i in range(c_matrix.shape[0])) / np.sum(c_matrix)
    return c_matrix, accuracy
```

```python
# cross validation
def cross_validation(X, y, c=1., sigma=0.5, degree=1, sov_lower=0.001, mode=None, \
                     kernel_func=None, n_folds=5, r_verbose=False):
    train_c_matrix = []
    train_accuracy = []
    
    test_c_matrix = []
    test_accuracy = []
    
    train_idx, test_idx = K_Fold(y.shape[0], n_folds=n_folds, shuffle=True)
    
    for i in range(n_folds):
        # fit
        alpha_sv, X_sv, y_sv, alpha, W, W_0 = fit(X, y, c=c, sigma=sigma, degree=degree, sov_lower=sov_lower, \
                                                  mode=mode, kernel_func=kernel_func)
        
        # training error
        train_predicted = predict(X[train_idx[i]], W, W_0)
        train_label = (y[train_idx[i]] + 1.) / 2.
        tr_c_matrix, tr_accuracy = score(train_predicted, train_label)

        # test error
        test_predicted = predict(X[test_idx[i]], W, W_0)
        test_label = (y[test_idx[i]] + 1.) / 2.
        t_c_matrix, t_accuracy = score(test_predicted, test_label)
        
        train_c_matrix.append(tr_c_matrix)
        train_accuracy.append(tr_accuracy)
        
        test_c_matrix.append(t_c_matrix)
        test_accuracy.append(t_accuracy)
        
    # average error
    train_accuracy_avg = np.mean(train_accuracy)
    
    test_accuracy_avg = np.mean(test_accuracy)

    if r_verbose:
        print '(average results) CROSS_VALIDATION (degree = %d) (n_folds = %d):' % (degree, n_folds)
        print 'train_c_matrix, train_accuracy_avg'
        for i in range(len(train_c_matrix)):
            print train_c_matrix[i]
        print train_accuracy_avg
        print '\n'
        print 'test_c_matrix, test_accuracy_avg'
        for i in range(len(test_c_matrix)):
            print test_c_matrix[i]
        print test_accuracy_avg
        print '\n'
```

# 5-2. Test the algorithm on two additional external datasets

```python
# read file
def read_file(fname, delimiter=None, converters=None):
    data = np.loadtxt(fname, delimiter=delimiter, converters=converters)
    return data
```

```python
def default_value(s):
    if s == 'Iris-setosa':
        return 1.
    elif s == 'Iris-versicolor':
        return 2.
    elif s == 'Iris-virginica':
        return 3.
    elif s == '?':
        return np.nan
    
    return s
```

```python
def normalization(X):
    for i in range(X.shape[1]):
        x = np.array(X[:,i])
        xmax = np.amax(X[:, i])
        xmin = np.amin(X[:, i])
        X[:, i] = (x - xmin) / (xmax - xmin)
    return X
```

```python
def convert_missing_value(X):
    col_mean = np.nanmean(X, axis=1)
    idx = np.where(np.isnan(X))
    X[idx] = np.take(col_mean, idx[1])
    return X
```

```python
# read file
iris_fname = 'iris.data'
iris_data = read_file(iris_fname, delimiter=',', converters={0: default_value, 1: default_value, \
                                                             2: default_value, 3: default_value, 4: default_value})
iris_s = np.split(iris_data, [iris_data.shape[1] - 1], axis=1)
iris_X23_0 = convert_missing_value(iris_s[0][50:100])
iris_X23_1 = convert_missing_value(iris_s[0][100:150])
iris_X23 = np.append(iris_X23_0, iris_X23_1, axis=0)
iris_y23 = (iris_s[1][50:150] * 2.) - 5.
iris_X23 = normalization(iris_X23)
print iris_X23
print iris_y23
iris_dataset = (iris_X23, iris_y23)
```

```python
# polynomial hernel function: iris_dataset
X, y = iris_dataset

c = 1
degree = 2
sov_lower = 0.0001
mode = 'soft'
kernel_func = 'poly'
n_folds = 5
r_verbose = True

# start timing
start = time.time()

# cross validation
cross_validation(X, y, c=c, degree=degree, sov_lower=sov_lower, mode=mode, \
                 kernel_func=kernel_func, n_folds=n_folds, r_verbose=r_verbose)

# stop timing
stop = time.time()
duration = stop - start

print 'duration =', duration

alpha_sv, X_sv, y_sv, alpha, W, W_0 = fit(X, y, c=c, sov_lower=sov_lower, degree=degree, \
                                          mode=mode, kernel_func=kernel_func)
predicted = predict(X, W, W_0)
label = (y + 1.) / 2.
c_matrix, accuracy = score(predicted, label)
print 'APPLY TO ALL DATA (degree = %d):' % degree
print 'c_matrix, accuracy'
print c_matrix
print accuracy
```
