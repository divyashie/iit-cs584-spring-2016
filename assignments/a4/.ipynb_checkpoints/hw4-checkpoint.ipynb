```python
def data2file(data, cls, file_name=None):
    if file_name == None:
        file_name = 'untitled.data'
    f = open(file_name, 'wb+')
    for iii in range(data.shape[0]):#70000
        for jjj in range(data.shape[1]):#784
            if jjj == range(data.shape[1] - 1):
                f.write(str(data[iii, jjj]))
            else:
                f.write(str(data[iii, jjj]) + ' ')
        f.write('\n')
    f.close()
    cls_file_name = file_name.rpartition('.')[0] + '_cls.data'
    f1 = open(cls_file_name, 'wb+')
    for i in range(len(cls)):#70000
        if i ==len(cls)-1:
            f1.write(str(cls[i]))
        else:
            f1.write(str(cls[i])+' ')
    f1.close()
    return 'Data Saved'
```

```python
def split_to_classes(X, y):
    cls = np.unique(y)
    feas_num = X.shape[1]
    idx_dict = {}
    buff = y==cls
    
    for k in range(buff.shape[1]):
        j = 0
        idx_t = []
        for i in buff[:, k]:
            if i == True:
                idx_t.append(j)
            j += 1
        idx_dict.update({cls[k]: idx_t})  
        
    return idx_dict
```

```python
def k_folds(x, n_folds=10):
    
    x_len = np.shape(x)[0]
    test_idx = []
    train_idx = []
    all_idx = range(x_len) 
    if n_folds != 1:
        for i in range(0, n_folds):
            t = all_idx[i*x_len//n_folds:(i+1)*x_len//n_folds]
            test_idx.append(t)
            train_idx.append(np.delete(all_idx, t))
            train_idx = test_idx
    else: 
        test_idx = [all_idx]
        train_idx = [all_idx]
 
    return test_idx, train_idx
```

```python
def data_shuffle(X, y, SEED=None):
    if SEED is None:
        SEED = np.random.randint(1, 999999)
    print '-=SEED=-'
    print SEED
    np.random.seed(SEED)
    np.random.shuffle(X)
    np.random.seed(SEED)
    np.random.shuffle(y)
    return X, y
```

```python
def cls_to_coln_int(y):
    if not isinstance(y, np.ndarray):
        y_arr = np.asarray(y)[:, np.newaxis]
    elif len(y.shape) == 1:
        y_arr = y[:, np.newaxis]
    elif y.shape[0] < y.shape[1]:
        y_arr = y.T
    else:
        y_arr = y
    
    y_t = np.zeros((y_arr.shape[0], 1), dtype = int)
    classes = np.unique(y_arr)
    ii = 0
    cls_vs_num = {}
    for cls in classes:
        y_t += ((y_arr==cls)+0) * ii
        cls_vs_num.update({cls: str(ii)})
        ii += 1
    
    return y_t, cls_vs_num
```

```python
def one_vs_all_other(y, cls, vs_val='ZERO'):
    """
        vs_val: 'ZERO' for '1 to 0'; 'NEG' for '1 to -1'
    """
    if vs_val is 'ZERO':
        y_vs = (y==cls)+0
    elif vs_val is 'NEG':
        y_vs = ((y==cls)+0)*2-1
    else:
        raise NameError('VS. Value :' + vs_val + 'is not defined.')
    return y_vs
```

```python
# For Training K(x_tr, x_tr)
# For Testing K(x_tr, x_tt)
def select_kernel(x1, x2, type_kernel='linear', c_poly=1, r_poly=2, sigma_Gaus=0.1):
    if type_kernel == 'poly':
        kernel = kernel_polynomial(x1=x1, x2=x2, c_poly=c_poly, r_poly=r_poly)
    elif type_kernel == 'Gaussian':
        kernel = kernel_Gaussian(x1=x1, x2=x2, sigma_Gaus=sigma_Gaus)
    elif type_kernel == 'linear':
        kernel = kernel_linear(x1=x1, x2=x2)
    else:
        raise NameError('Kernel Type: ' + type_kernel + 'is not defined.')
    return kernel, type_kernel
```

```python
def kernel_linear(x1, x2):
    return np.dot(x1, x2.T)
```

```python
def kernel_polynomial(x1, x2, c_poly=1, r_poly=2):
    return (np.dot(x1, x2.T) + c_poly) ** r_poly
```

```python
def kernel_Gaussian(x1, x2, sigma_Gaus):
    M_dist_euc = sp.spatial.distance.cdist(x1, x2, 'euclidean')
    return np.exp(-M_dist_euc**2/(2.*sigma_Gaus**2))
```

```python
# Lagrange_multiplier: alpha. To predict
def Lagrange_multiplier(kernel_tr, y, soft_mgn_val=None, b_Lagr='ZERO'):
    
#     if soft_mgn_val is None:
#         C = np.inf
#     elif not (isinstance(soft_mgn_val, np.float) or isinstance(soft_mgn_val, np.int)):
#         raise TypeError('soft_mgn_val is Float OR None!')
#     elif soft_mgn_val < 0:
#         raise ValueError('soft_mgn_val is Positive!')
#     else:
#         C = soft_mgn_val
    
    if b_Lagr is 'ZERO':
        b_new = 0.
    else:
        print '"b" To Be Updated! Assume as 0 now!'
        b_new = b_Lagr
        
    y_shape = y.shape
#     print y_shape
#     print np.dot(y,y.T)
    P = np.dot(y, y.T) * kernel_tr
#     P = np.dot(np.dot(y, y.T), kernel_tr.T)
    q = -np.ones(y_shape)
    
    if soft_mgn_val is None:
        G = -1. * np.identity(y_shape[0])
        h = 1.* np.zeros(y_shape)
    else: 
        C = 1.* soft_mgn_val
        G = np.vstack((-np.identity(y_shape[0]), np.identity(y_shape[0])))
        h = np.vstack((np.zeros(y_shape), C * np.ones(y_shape)))
    
    A = 1.*y.T
    
    
#     print X.shape
#     print X
#     print y.shape
#     print y
#     print'============'
#     print 'P', P.shape
#     print'------------'
#     print P
#     print'============'
#     print 'q', q.shape
#     print'------------'
#     print q
#     print'============'
#     print 'G', G.shape
#     print'------------'
#     print G
#     print'============'
#     print 'h', h.shape
#     print'------------'
#     print h
#     print'============'
#     print 'A', A.shape
#     print'------------'
#     print A
#     print'============'


#     print '----===Lagrange_multiplier->A===----' # [checked]-- [1;-1]
#     print A
#     print '----===Lagrange_multiplier->A--- END---===----'
    
    sol = (cv.solvers.qp(cv.matrix(P), cv.matrix(q), \
                         cv.matrix(G), cv.matrix(h), \
                         cv.matrix(A), cv.matrix(1.* b_new)))
#     sol = (cv.solvers.qp(cv.matrix(P), cv.matrix(q), \
#                          cv.matrix(G), cv.matrix(h)))['x']
    
#     print sol
#     print sol['x']
    return np.array(sol['x'])
```

```python
def functional_margin(x, y, alpha, sv_low_bound=0.1): # w: (1, n)
#     print 'x shape', x.shape
#     print 'y shape', y.shape
#     print 'al shape', alpha.shape
    w = (np.sum(np.dot((alpha * y).T, x), axis = 0))[:, np.newaxis]
#     p = np.sum((np.dot((alpha * y_tr).T, kernel_tt) + b), axis=0)
    print 'w shape', w.shape
#     print type(w)
    
    sv_idx = (alpha > sv_low_bound).T[0]
#     print np.sum((alpha > sv_low_bound)+0)
#     print alpha
    print 'sv_idx shape', sv_idx
#     print x[sv_idx]
    b = y[sv_idx] - np.dot(x[sv_idx], w)
    print '== b ==='
    print b
    print '====='
    return w, np.mean(b), x[sv_idx]
```

```python
def predict(kernel_tt, y_tr, alpha, b): # 1X1
#     print
    pdt = np.sum((np.dot((alpha * y_tr).T, kernel_tt) + b), axis=0)
#     print 'predict.pdt'
#     print pdt[:, np.newaxis]
    return pdt[:, np.newaxis]
```

```python
def classify_SVM(kernel_t, y_t, alpha, b):
    pdts = {}
#     print '+++++'
#     print y_t.values()[0] + y_t.values()[1]
#     print '+++++'
#     print '+++++'
#     print alpha.values()[0] - alpha.values()[1]
#     print '===='
#     print alpha.values()[1]
#     print '+++++'
    for cls in y_t.keys():
#         print y_t.get(cls)
        pdt = predict(kernel_t, y_t.get(cls), alpha.get(cls), b.get(cls))
#         print pdt.shape
        pdts.update({cls: pdt})
    
    
    
#     print pdts.values()[0].shape
    idx_max = np.argmax(np.hstack(pdts.values()), axis=1)
    
#     print idx_max
    cls_pdt = [pdts.keys()[i] for i in idx_max]

    return np.asarray(cls_pdt)
```

```python
def confusion_matrix(y_true, y_pdt, labels=None):
    
    if labels is None:
        labels = np.unique(y_true)
    else:
        labels = np.asarray(labels)
    
    n_labels = labels.shape[0]
    
    label_to_idx = dict((y, x) for x, y in enumerate(labels))
    y_pdt = np.array([label_to_idx.get(x, n_labels + 1) for x in y_pdt])
    
    y_true = np.array([label_to_idx.get(x, n_labels + 1) for x in y_true])
    
    idx = np.logical_and(y_pdt < n_labels, y_true < n_labels)
    
    y_pdt = y_pdt[idx]
    y_true = y_true[idx]
    
    cfs_mtrx = coo_matrix((np.ones(y_true.shape[0], dtype=np.int), \
                           (y_true, y_pdt)), \
                          shape=(n_labels, n_labels)).toarray()
    
    return cfs_mtrx, labels
```

```python
def precision_recall_fscore(y_true, y_pdt, beta=1):
    
    cm, labels= confusion_matrix(y_true, y_pdt)
    if beta <= 0 or isinstance(beta, np.complex):
        raise ValueError("Beta is non-negative real value.")
    
    total = np.sum(cm)
    psns_t = []
    rcls_t = []
    fscs_t = []
    acrs_t = []
    
    for i in range(len(labels)):
        tp = 1.* cm[i, i]
        fn_tp = 1. * np.sum(cm[i, :])
        fp_tp = 1. * np.sum(cm[:, i])
        tn = total - fn_tp - fp_tp + tp

        psn_t = tp/fp_tp
        rcl_t = tp/fn_tp
        fsc_t = (1.+beta**2)*psn_t*rcl_t/(beta**2*psn_t+rcl_t)
        acr_t = (tp+tn)/total
        
        psns_t.append(psn_t)
        rcls_t.append(rcl_t)
        fscs_t.append(fsc_t)
        acrs_t.append(acr_t)
        
        
    return psns_t, rcls_t, fscs_t, acrs_t, labels
```

```python
def score(y_true, y_pdt):
    corrct = np.sum((y_pdt==y_true)+0)# predict correct
    acc = 1.*corrct/len(y_true)
    return acc, 1-acc
```

```python
def cross_validation_SVM(X, y, shuffle=True, SEED=None, normalized = False, vs_val='NEG', \
                         n_folds=5, type_kernel='linear', c_poly=1, r_poly=2, sigma_Gaus=0.1, \
                         soft_mgn_val=0, b_Lagr='ZERO', sv_low_bound=0.1, verbose=True):

    if shuffle:
        X, y_temp = data_shuffle(X, y, SEED=SEED)
    else: 
        X, y_temp = X, y
        
    if normalized is False:
        print 'DATA IS NOT NORMALIZED!'
    else: 
        print 'NORMALIZING DATA IS DEVELOPING!'
        
    test_idx, train_idx = k_folds(X, n_folds=n_folds)
        
    y, cls_vs_num = cls_to_coln_int(y_temp)
    
    err_tts = []
    err_trs = []
    psns = []
    rcls = []
    fscs = []
    acrs = []
    
    for i in range(n_folds):
        # train
        y_train = y[train_idx[i]]
        X_train = X[train_idx[i]]
        
#         fig1 = plt.figure(1)
#         plt.scatter(X_train[:, 0], X_train[:, 1], marker='^', c=y_train)
#         fig1.show()
#         plt.axis([-1, 3.5, -3, 3])
        
        cls_train = np.unique(y_train)
        
        cls_alpha = {}
        cls_w = {}
        cls_b = {}
        sv = []
        cls_y_one_all_tr = {}
        
        kernel_tr, _ = select_kernel(x1=X_train, x2=X_train, type_kernel=type_kernel, \
                                              c_poly=c_poly, r_poly=r_poly, sigma_Gaus=sigma_Gaus)
#         print kernel_tr.shape
        '''
        # Main problem is if train or test in cls and if YES, what value should be passed in as cls
        # A: train: per cls
        #    test: per cls and pick up Max score among classes: <w_tr.T,x_tt> + b
        #          w_tr = alpha_tr*y_tr*x_tr, <w_tr.T, x_tt> = alpha_tr * y_tr * K(x_tr, x_tt)
        '''
        
        for cls in cls_train:
            print 'cls =', cls
            y_train_one2all = one_vs_all_other(y=y_train, cls=cls, vs_val=vs_val) # 'NEG': '1 to -1'
            
#             print '----===y_train_one2all===----'[check]-- [1,-1; -1,1]
#             print y_train_one2all
#             print '----===y_train_one2all--- END---===----'
#             print '----===Kernel_tr===----'[check]-- ALL SAME
#             print kernel_tr
#             print '----===Kernel_tr--- END---===----'

            alpha_tr = Lagrange_multiplier(kernel_tr, y_train_one2all, soft_mgn_val=soft_mgn_val, b_Lagr=b_Lagr)
            
#             print '----===alpha_tr===----'
#             print alpha_tr
#             print '----===alpha_tr--- END---===----'
            
            w_tr, b_tr, sv_tr = functional_margin(X_train, y_train_one2all, alpha_tr, sv_low_bound)
            
            cls_alpha.update({cls: alpha_tr})
            cls_w.update({cls: w_tr})
            cls_b.update({cls: b_tr})
            sv.append(sv_tr)
            cls_y_one_all_tr.update({cls: y_train_one2all})
            #####
            #plot
            ####
#             plot_boundary()
         
        #sv = X_train[sv_idx]
#         print sv

#         fig2 = plt.figure()
#         x_min = X_train[:, 0].min()
#         x_max = X_train[:, 0].max()
#         y_min = X_train[:, 1].min()
#         y_max = X_train[:, 1].max()
#         XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
#         Z = 2 * YY - XX
#         plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
#         plt.contour(XX, YY, Z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'], levels=[-.5, 0, .5])
#         plt.scatter(sv[0][:, 0], sv[0][:, 1], s=80, facecolors='none', zorder=10)
#         plt.axis('tight')
#         plt.scatter(X_train[:, 0], X_train[:, 1], marker='o', c=y_train, zorder=10, cmap=plt.cm.Paired)
#         fig2.show()
#         plt.axis([-1, 3.5, -3, 3])



        print 'cls_alpha =', cls_alpha
        print 'cls_w =', cls_w
        print 'cls_b =', cls_b
        print 'cls_y_one_all_tr =', cls_y_one_all_tr
        # train error
        cls_pdt_tr = classify_SVM(kernel_tr, cls_y_one_all_tr, cls_alpha, cls_b)
        cls_true_tr = y_train[:, 0]
        #  
        _, err_tr = score(cls_true_tr, cls_pdt_tr)
            
            
        # test
        y_test = y[test_idx[i]]
        X_test = X[test_idx[i]]


        '''
        # x1: X_Train or X_SV=X_Train[sv_idx]
        '''
        # kernel_tt: K(x_tr, x_tt)
        kernel_tt, _ = select_kernel(x1=X_train, x2=X_test, type_kernel=type_kernel, \
                                              c_poly=c_poly, r_poly=r_poly, sigma_Gaus=sigma_Gaus)
            
        # test error
        '''
        # TEST: <w_tr.T, x_test> + b -> pick up Max one among each classes
        # w_tr = alpha_tr*y_tr*x_tr
        # <w_tr.T, x_tt> = alpha_tr * y_tr * K(x_tr, x_tt)
        '''
        
        cls_pdt_tt = classify_SVM(kernel_tt, cls_y_one_all_tr, cls_alpha, cls_b) 
        cls_true_tt = np.asarray(y_test[:, 0])
        
        ####### for testing
#         print 'pdt_cls1', cls_pdt_tt
        
        _, err_tt = score(cls_pdt_tt, cls_true_tt)
        
        err_trs.append(err_tr)
        err_tts.append(err_tt)
        
        conf_matrix, labels = confusion_matrix(cls_true_tt, cls_pdt_tt)
            
        if verbose:
            print 'folder: %d' % i
            print labels
            print '-=training error=-'
            print err_tr
            print '-=testing error=-'
            print err_tt
            print '-=confusion matrix=-'
            print conf_matrix
    
        psn, \
        rcl, \
        fsc, \
        acr, _ = precision_recall_fscore(cls_true_tt, cls_pdt_tt)
        
        psns.append(psn)
        rcls.append(rcl)
        fscs.append(fsc)
        acrs.append(acr)

    precision = np.asarray(psns)
    recall = np.asarray(rcls)
    f_score = np.asarray(fscs)
    accuracy = np.asarray(acrs)
    
    if verbose:
        print '-=labels=-'
        print labels
        print '-=precision=-'
        print precision
        print '-=recall=-'
        print recall
        print '-=f_score=-'
        print f_score
        print '-=accuracy=-'
        print accuracy
```
